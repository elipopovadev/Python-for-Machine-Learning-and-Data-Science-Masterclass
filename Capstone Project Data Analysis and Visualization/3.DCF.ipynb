# 2. Inputs for DCF

```python
##### print('--Inputs for DCF--')
n = int(input('Enter the number of years: '))  # 5
t = float(input('Enter the effective tax rate (in %): '))  # 26
g = float(input('Enter the growth rate (in %): '))  # 3 
wacc = float(input('Enter the WACC (in %): ')) # 6 
ebit = float(EBIT)
DA = float(DEPRECIATION_AMORTIZATION)
capex = float(CAPITAL_EXPENDITURES)
cWC = float(CHANGE_WORKING_CAPITAL)
```

# 3. DCF

```python
print('--Forecast EBIT and Tax--')
current_ebit = ebit  # Store initial EBIT (from financials)
L1 = []
L2 = []
print(f'EBIT for current year (Year 0) is {round(current_ebit, 2)}')
for i in range(1, n + 1):
    forecast_ebit = current_ebit * (1 + g/100) ** i  # Apply growth ONCE per year
    L1.append(forecast_ebit)
    
    tax = forecast_ebit * (t / 100)
    L2.append(tax)
    
    print(f'1] EBIT for forecast year {i}: {round(forecast_ebit, 2)}')
    print(f'2] Tax for forecast year {i}: {round(tax, 2)}')
print('----------------------------------------------------------------------------------------')   
```

```python
print('--Forecast D & A--')
current_da = DA  # Store initial D&A (from financials)
L3 = [current_da]  # Initialize with current year
print(f'D&A for current year (Year 0) is {round(current_da, 2)}')
for i in range(1, n+1):  # Start from Year 1 (skip Year 0)
    forecast_da = current_da * (1 + g/100) ** i  # Apply growth
    L3.append(forecast_da)
    print(f'3] D&A for forecast year {i} is {round(forecast_da, 2)}')
print('----------------------------------------------------------------------------------------')    
```

```python
print('--Forecast Capex--')
current_capex = capex  # Store initial Capex (from financials)
L4 = [current_capex]  # Initialize with current year
print(f'Capex for current year (Year 0) is {round(current_capex, 2)}')

for i in range(1, n+1):  # Start from Year 1 (skip Year 0)
    forecast_capex = current_capex * (1 + g/100) ** i  # Apply growth
    L4.append(forecast_capex)
    print(f'4] Capex for forecast year {i} is {round(forecast_capex, 2)}')
print('----------------------------------------------------------------------------------------')    
```

```python
print('--Forecast WC--')
current_WC = cWC   # Current WC (Year 0)
print(f'Current WC (Year 0) is {round(current_WC, 2)}')
L5 = []
prev_WC = current_WC 
for i in range(1, n+1):
    forecast_WC = current_WC   * ((1 + g/100) ** i)
    # WC = WC_year_i - WC_year_i-1
    delta_WC = forecast_WC - prev_WC
    L5.append(delta_WC)
    prev_WC = forecast_WC
    print(f'WC Year {i}: {round(delta_WC, 2)}')
print('----------------------------------------------------------------------------------------')
```

```python
print('--Forecast FCF--')
# Start with actual Year 0 FCF
FCF = [cash_flow.loc['Free Cash Flow'].iloc[0]]
print(f'FCF for current year (Year 0) is {round(FCF[0], 2)}')

# Forecast Years 1 to n
for i in range(n):
    fcf = (L1[i] - L2[i]) + L3[i] - L4[i] + L5[i]
    FCF.append(fcf)
    print(f'FCF for forecast year {i + 1} is {round(fcf, 2)}')

# Terminal Value at end of Year n
last_projected_fcf = FCF[-1]  # This is Year n FCF
if wacc/100 <= g/100:
    raise ValueError("WACC must be greater than the growth rate (g)")
TV = (last_projected_fcf * (1 + g/100)) / (wacc/100 - g/100)
print(f'Terminal Value at year {n}: {round(TV, 2)}')
print('----------------------------------------------------------------------------------------')
```

```python
##### Present Value Calculations

PV_FCF = []

# Discount FCF from year 1 to n (FCF[1] to FCF[n])
for i in range(1, n + 1):
    pv = FCF[i] / ((1 + wacc / 100) ** i)
    PV_FCF.append(pv)
    print(f'Present Value of FCF for year {i}: {round(pv, 2)}')

# Discount Terminal Value (at Year n) back to present
PV_TV = TV / ((1 + wacc / 100) ** n)
print(f'Present Value of Terminal Value: {round(PV_TV, 2)}')

# Enterprise Value = PV of FCF (years 1 to n) + PV of Terminal Value
EV = sum(PV_FCF) + PV_TV
print(f'Enterprise Value (EV): {round(EV, 2)}')
print('----------------------------------------------------------------------------------------')
```

# 4. Intrinsic Value 

```python
cash = float(balance_sheet_df.loc['Cash And Cash Equivalents'].iloc[0])
Sdbt = float(balance_sheet_df.loc['Current Debt'].iloc[0])
Ldbt = float(balance_sheet_df.loc['Long Term Debt'].iloc[0])
```

```python
EOV = EV - (Sdbt + Ldbt) + cash
print('The Equity Value (Eq.V) is ', round(EOV))
```

```python
diluted_shares = income_statement.loc["Diluted Average Shares"].iloc[0]
diluted_shares
```

```python
IV = EOV / diluted_shares
print('The Intrinsic Value (IV) is ', round(IV,2))
```

__Standard P/B Ratio & Adjusted Price to Book__

```python
market_cap = float(stock_info['marketCap'])
total_equity = float(balance_sheet_df.loc['Stockholders Equity'].iloc[0])
intangible_assets = float(balance_sheet_df.loc['Goodwill And Other Intangible Assets'].iloc[0])

# Standard book value
book_value = total_equity
book_value_per_share = book_value / diluted_shares  
market_price_per_share = market_cap / diluted_shares  

# Standard P/B Ratio
pb_ratio = market_price_per_share / book_value_per_share

# Adjusted Book Value (remove intangibles)
adjusted_book_value = total_equity - intangible_assets
adjusted_pb_ratio = market_cap / adjusted_book_value

print(f"Standard P/B Ratio: {pb_ratio:.2f}")
print(f"Adjusted P/B Ratio: {adjusted_pb_ratio:.2f}")
```
